```python
# Separar TDM de controles
controles = []
tdm = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]
    if 'Control' in titulo:
        controles.append(nombre)
    elif 'MDD' in titulo or 'Depression' in titulo or 'Case' in titulo:
        tdm.append(nombre)

print(f"Total muestras: {len(gse.gsms)}")
print(f"Controles sanos: {len(controles)}")
print(f"Pacientes TDM: {len(tdm)}")
print(f"\nPrimeros 3 controles: {controles[:3]}")
print(f"Primeros 3 TDM: {tdm[:3]}")
```

```python
import pandas as pd

# Crear tabla con información de cada muestra
info_muestras = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]

    # Determinar grupo
    if 'Control' in titulo:
        grupo = 'Control'
    else:
        grupo = 'TDM'

    # Determinar sexo
    if '-M-' in titulo:
        sexo = 'Masculino'
    elif '-F-' in titulo:
        sexo = 'Femenino'
    else:
        sexo = 'Desconocido'

    info_muestras.append({
        'ID_muestra': nombre,
        'Titulo': titulo,
        'Grupo': grupo,
        'Sexo': sexo
    })

# Convertir a tabla
df_muestras = pd.DataFrame(info_muestras)

# Mostrar resumen
print("=== RESUMEN DEL DATASET GSE251778 ===\n")
print(df_muestras.groupby(['Grupo', 'Sexo']).size())
print(f"\nTotal: {len(df_muestras)} muestras")

# Guardar como archivo CSV
df_muestras.to_csv('metadata_GSE251778.csv', index=False)
print("\n✓ Metadata guardada en metadata_GSE251778.csv")
```

```python
import pandas as pd

# Cargar la matriz de expresión génica
print("Cargando matriz de expresión génica...")
df_expresion = pd.read_csv(
    './datos/GSE251778/GSE251778_Mokhtari_rna_counts.csv.gz',
    compression='gzip',
    index_col=0
)

print(f"✓ Matriz cargada")
print(f"Genes totales: {df_expresion.shape[0]}")
print(f"Muestras totales: {df_expresion.shape[1]}")
print(f"\nPrimeros 5 genes:")
print(df_expresion.head())
```

```python
# Buscar IDO1 en los datos
# IDO1 tiene el código Ensembl: ENSG00000131203

print("=== BUSCANDO IDO1 ===\n")

# Buscar por código Ensembl
ido1_ensembl = 'ENSG00000131203'

if ido1_ensembl in df_expresion['gene_ID'].values:
    ido1_data = df_expresion[
        df_expresion['gene_ID'] == ido1_ensembl
    ]
    print("✓ IDO1 ENCONTRADO EN EL DATASET")
    print(f"Código Ensembl: {ido1_ensembl}")
    print(f"\nNiveles de expresión en primeras 10 muestras:")
    print(ido1_data.iloc[:, 1:11])

    # Estadísticas básicas
    valores = ido1_data.iloc[:, 1:].values.flatten()
    print(f"\nExpresión mínima: {valores.min()}")
    print(f"Expresión máxima: {valores.max()}")
    print(f"Expresión promedio: {valores.mean():.2f}")
else:
    print("IDO1 no encontrado con ese código")
    print("Buscando variantes...")
    resultados = df_expresion[
        df_expresion['gene_ID'].str.contains('ENSG00000131203',
                                              na=False)
    ]
    print(resultados)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extraer valores de IDO1
ido1_valores = df_expresion[
    df_expresion['gene_ID'] == 'ENSG00000131203'
].iloc[:, 1:]

# Identificar columnas de cada grupo
# Los IDs de muestra tienen el formato: 111_rna, 113_rna, etc.
muestras_cols = ido1_valores.columns.tolist()

# Separar por grupo usando el metadata que guardamos
tdm_ids = df_muestras[df_muestras['Grupo'] == 'TDM']['ID_muestra'].tolist()
ctrl_ids = df_muestras[df_muestras['Grupo'] == 'Control']['ID_muestra'].tolist()

# Obtener valores numéricos de IDO1
ido1_fila = ido1_valores.values.flatten()

# Calcular promedio por grupo (aproximación con posiciones)
n_total = len(ido1_fila)
mitad = n_total // 2

grupo1 = ido1_fila[:mitad]
grupo2 = ido1_fila[mitad:]

print("=== IDO1: COMPARACIÓN ENTRE GRUPOS ===\n")
print(f"Grupo 1 — Promedio IDO1: {grupo1.mean():.2f}")
print(f"Grupo 2 — Promedio IDO1: {grupo2.mean():.2f}")
print(f"Diferencia: {abs(grupo1.mean() - grupo2.mean()):.2f}")

# Gráfico
fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot([grupo1, grupo2],
           labels=['Grupo 1', 'Grupo 2'])
ax.set_title('Expresión de IDO1 entre grupos\nDataset GSE251778')
ax.set_ylabel('Nivel de expresión (conteos)')
ax.set_xlabel('Grupo')
plt.tight_layout()
plt.savefig('IDO1_expresion_grupos.png', dpi=150)
plt.show()
print("\n✓ Gráfico guardado")
```

```python
# Conectar IDs del dataset con diagnóstico real
print("=== CONECTANDO IDs CON DIAGNÓSTICO ===\n")

# Ver cómo son los IDs en la matriz de expresión
print("IDs en matriz de expresión (primeros 10):")
print(df_expresion.columns[1:11].tolist())

print("\nIDs en metadata (primeros 10):")
print(df_muestras['ID_muestra'].head(10).tolist())

print("\nTítulos en metadata (primeros 5):")
print(df_muestras['Titulo'].head().tolist())
```

```python
# Extraer número del título para hacer el puente
print("=== MAPEANDO IDs CORRECTAMENTE ===\n")

# Extraer número del título: "WholeBlood-Control-M-101" → "101"
df_muestras['numero'] = df_muestras['Titulo'].str.extract(r'(\d+)$')

# Extraer número del ID de la matriz: "111_rna" → "111"
ids_matriz = pd.Series(df_expresion.columns[1:].tolist())
numeros_matriz = ids_matriz.str.extract(r'^(\d+)')[0]

print("Números en metadata (primeros 5):")
print(df_muestras['numero'].head().tolist())

print("\nNúmeros en matriz (primeros 5):")
print(numeros_matriz.head().tolist())

# Crear diccionario de mapeo número → grupo
mapa_grupos = dict(zip(
    df_muestras['numero'],
    df_muestras['Grupo']
))

# Asignar grupo a cada columna de la matriz
grupos_columnas = []
for num in numeros_matriz:
    grupo = mapa_grupos.get(num, 'Desconocido')
    grupos_columnas.append(grupo)

# Contar
from collections import Counter
conteo = Counter(grupos_columnas)
print("\nColumnas identificadas:")
print(f"TDM: {conteo['TDM']}")
print(f"Control: {conteo['Control']}")
print(f"Desconocido: {conteo['Desconocido']}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Extraer valores de IDO1
ido1_fila = df_expresion[
    df_expresion['gene_ID'] == 'ENSG00000131203'
].iloc[:, 1:].values.flatten()

# Separar por grupo real
ido1_tdm = []
ido1_ctrl = []

for i, grupo in enumerate(grupos_columnas):
    if grupo == 'TDM':
        ido1_tdm.append(ido1_fila[i])
    elif grupo == 'Control':
        ido1_ctrl.append(ido1_fila[i])

ido1_tdm = np.array(ido1_tdm)
ido1_ctrl = np.array(ido1_ctrl)

# Estadísticas
print("=== IDO1: TDM vs CONTROL ===\n")
print(f"TDM     — Promedio: {ido1_tdm.mean():.1f} | "
      f"Mediana: {np.median(ido1_tdm):.1f}")
print(f"Control — Promedio: {ido1_ctrl.mean():.1f} | "
      f"Mediana: {np.median(ido1_ctrl):.1f}")

# Test estadístico
stat, pvalor = stats.mannwhitneyu(ido1_tdm, ido1_ctrl,
                                   alternative='two-sided')
print(f"\nTest Mann-Whitney U:")
print(f"p-valor: {pvalor:.4f}")
if pvalor < 0.05:
    print("✓ Diferencia SIGNIFICATIVA entre grupos (p < 0.05)")
else:
    print("→ Diferencia no significativa (p ≥ 0.05)")

# Gráfico
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot
axes[0].boxplot([ido1_ctrl, ido1_tdm],
                tick_labels=['Control\n(n=89)', 'TDM\n(n=80)'],
                patch_artist=True,
                boxprops=dict(facecolor='lightblue'),)
axes[0].set_title('Expresión de IDO1\nControl vs TDM')
axes[0].set_ylabel('Conteos de expresión')

# Agregar puntos individuales
axes[0].scatter([1]*len(ido1_ctrl), ido1_ctrl,
                alpha=0.3, color='steelblue', s=15)
axes[0].scatter([2]*len(ido1_tdm), ido1_tdm,
                alpha=0.3, color='coral', s=15)

# Anotar p-valor
axes[0].text(1.5, max(max(ido1_ctrl), max(ido1_tdm))*0.95,
             f'p = {pvalor:.4f}'
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Extraer valores de IDO1
ido1_fila = df_expresion[
    df_expresion['gene_ID'] == 'ENSG00000131203'
].iloc[:, 1:].values.flatten()

# Separar por grupo real
ido1_tdm = []
ido1_ctrl = []

for i, grupo in enumerate(grupos_columnas):
    if grupo == 'TDM':
        ido1_tdm.append(ido1_fila[i])
    elif grupo == 'Control':
        ido1_ctrl.append(ido1_fila[i])

ido1_tdm = np.array(ido1_tdm)
ido1_ctrl = np.array(ido1_ctrl)

# Estadísticas
print("=== IDO1: TDM vs CONTROL ===\n")
print(f"TDM     — Promedio: {ido1_tdm.mean():.1f} | Mediana: {np.median(ido1_tdm):.1f}")
print(f"Control — Promedio: {ido1_ctrl.mean():.1f} | Mediana: {np.median(ido1_ctrl):.1f}")

# Test estadístico
stat, pvalor = stats.mannwhitneyu(ido1_tdm, ido1_ctrl, alternative='two-sided')
print(f"\nTest Mann-Whitney U:")
print(f"p-valor: {pvalor:.4f}")
if pvalor < 0.05:
    print("DIFERENCIA SIGNIFICATIVA entre grupos (p < 0.05)")
else:
    print("Diferencia no significativa (p >= 0.05)")

# Gráfico
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot
axes[0].boxplot([ido1_ctrl, ido1_tdm],
                tick_labels=['Control\n(n=89)', 'TDM\n(n=80)'],
                patch_artist=True,
                boxprops=dict(facecolor='lightblue'))
axes[0].set_title('Expresion de IDO1\nControl vs TDM')
axes[0].set_ylabel('Conteos de expresion')
axes[0].scatter([1]*len(ido1_ctrl), ido1_ctrl, alpha=0.3, color='steelblue', s=15)
axes[0].scatter([2]*len(ido1_tdm), ido1_tdm, alpha=0.3, color='coral', s=15)
axes[0].text(1.5, max(max(ido1_ctrl), max(ido1_tdm))*0.95,
             f'p = {pvalor:.4f}', ha='center', fontsize=11, fontweight='bold')

# Histograma
axes[1].hist(ido1_ctrl, bins=25, alpha=0.6, color='steelblue', label='Control')
axes[1].hist(ido1_tdm, bins=25, alpha=0.6, color='coral', label='TDM')
axes[1].set_title('Distribucion de IDO1\npor grupo')
axes[1].set_xlabel('Conteos de expresion')
axes[1].set_ylabel('Numero de muestras')
axes[1].legend()

plt.suptitle('IDO1 (ENSG00000131203) - Dataset GSE251778', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('IDO1_TDM_vs_Control.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nGrafico guardado: IDO1_TDM_vs_Control.png")
```
