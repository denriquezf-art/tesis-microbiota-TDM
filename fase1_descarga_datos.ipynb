```python
# Separar muestras MDD vs Control
mdd = []
control = []

for nombre, muestra in gse.gsms.items():
    chars = muestra.metadata.get('characteristics_ch1', [])
    chars_str = ' '.join(chars).lower()

    if 'mdd' in chars_str or 'depressiv' in chars_str:
        mdd.append(nombre)
    elif 'control' in chars_str:
        control.append(nombre)

print(f"Muestras MDD:     {len(mdd)}")
print(f"Muestras Control: {len(control)}")
print(f"Total:            {len(mdd) + len(control)}")
print("\nEjemplos MDD:", mdd[:3])
print("Ejemplos Control:", control[:3])
```

```python
import matplotlib.pyplot as plt

# Gráfico de distribución de muestras
grupos = ['MDD', 'Control']
cantidades = [len(mdd), len(control)]
colores = ['#D85A30', '#02A896']

plt.figure(figsize=(7, 5))
bars = plt.bar(grupos, cantidades, color=colores,
               width=0.5, edgecolor='white', linewidth=1.5)

for bar, cant in zip(bars, cantidades):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 1,
             str(cant), ha='center',
             fontsize=14, fontweight='bold')

plt.title('Dataset GSE251778\nDistribución de muestras',
          fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Número de muestras', fontsize=12)
plt.ylim(0, 110)
plt.tight_layout()
plt.savefig('distribucion_muestras.png', dpi=150)
plt.show()
print("✓ Gráfico guardado")
```

```python
import pandas as pd

# Construir tabla de metadatos clínicos
metadatos = []

for nombre, muestra in gse.gsms.items():
    chars = muestra.metadata.get('characteristics_ch1', [])
    fila = {'sample_id': nombre}
    for c in chars:
        if ':' in c:
            clave, valor = c.split(':', 1)
            fila[clave.strip()] = valor.strip()
    metadatos.append(fila)

df_meta = pd.DataFrame(metadatos)
print("Tabla de metadatos:")
print(df_meta.head(10))
print(f"\nColumnas disponibles: {list(df_meta.columns)}")
df_meta.to_csv('metadatos_GSE251778.csv', index=False)
print("✓ Tabla guardada como CSV")
```

```python
# Verificar distribución en la tabla
print("=== DISTRIBUCIÓN POR GRUPO ===")
print(df_meta['group'].value_counts())

print("\n=== DISTRIBUCIÓN POR SEXO ===")
print(df_meta['Sex'].value_counts())

print("\n=== MDD por sexo ===")
print(df_meta.groupby(['group', 'Sex']).size())
```

```python
# FASE 1 — Descarga del dataset de transcriptómica
import GEOparse

print("Conectando con NCBI GEO...")
gse = GEOparse.get_GEO(geo="GSE251778",
                        destdir="./datos/",
                        silent=False)

print("\n--- RESULTADO ---")
print("Título del estudio:", gse.metadata['title'][0])
print("Número de muestras:", len(gse.gsms))
```

```python
# Ver los primeros 5 pacientes del dataset
print("Primeras 5 muestras del estudio:\n")

for i, (nombre, muestra) in enumerate(gse.gsms.items()):
    if i >= 5:
        break
    titulo = muestra.metadata.get('title', ['Sin título'])[0]
    caracteristica = muestra.metadata.get(
        'characteristics_ch1', ['Sin info'])
    print(f"ID: {nombre}")
    print(f"Título: {titulo}")
    print(f"Características: {caracteristica[0]}")
    print("---")
```

```python
# Separar TDM de controles
controles = []
tdm = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]
    if 'Control' in titulo:
        controles.append(nombre)
    elif 'MDD' in titulo or 'Depression' in titulo or 'Case' in titulo:
        tdm.append(nombre)

print(f"Total muestras: {len(gse.gsms)}")
print(f"Controles sanos: {len(controles)}")
print(f"Pacientes TDM: {len(tdm)}")
print(f"\nPrimeros 3 controles: {controles[:3]}")
print(f"Primeros 3 TDM: {tdm[:3]}")
```

```python
import pandas as pd

# Crear tabla con información de cada muestra
info_muestras = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]

    # Determinar grupo
    if 'Control' in titulo:
        grupo = 'Control'
    else:
        grupo = 'TDM'

    # Determinar sexo
    if '-M-' in titulo:
        sexo = 'Masculino'
    elif '-F-' in titulo:
        sexo = 'Femenino'
    else:
        sexo = 'Desconocido'

    info_muestras.append({
        'ID_muestra': nombre,
        'Titulo': titulo,
        'Grupo': grupo,
        'Sexo': sexo
    })

# Convertir a tabla
df_muestras = pd.DataFrame(info_muestras)

# Mostrar resumen
print("=== RESUMEN DEL DATASET GSE251778 ===\n")
print(df_muestras.groupby(['Grupo', 'Sexo']).size())
print(f"\nTotal: {len(df_muestras)} muestras")

# Guardar como archivo CSV
df_muestras.to_csv('metadata_GSE251778.csv', index=False)
print("\n✓ Metadata guardada en metadata_GSE251778.csv")
```
