```python
# Separar muestras MDD vs Control
mdd = []
control = []

for nombre, muestra in gse.gsms.items():
    chars = muestra.metadata.get('characteristics_ch1', [])
    chars_str = ' '.join(chars).lower()

    if 'mdd' in chars_str or 'depressiv' in chars_str:
        mdd.append(nombre)
    elif 'control' in chars_str:
        control.append(nombre)

print(f"Muestras MDD:     {len(mdd)}")
print(f"Muestras Control: {len(control)}")
print(f"Total:            {len(mdd) + len(control)}")
print("\nEjemplos MDD:", mdd[:3])
print("Ejemplos Control:", control[:3])
```

```python
import matplotlib.pyplot as plt

# Gráfico de distribución de muestras
grupos = ['MDD', 'Control']
cantidades = [len(mdd), len(control)]
colores = ['#D85A30', '#02A896']

plt.figure(figsize=(7, 5))
bars = plt.bar(grupos, cantidades, color=colores,
               width=0.5, edgecolor='white', linewidth=1.5)

for bar, cant in zip(bars, cantidades):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 1,
             str(cant), ha='center',
             fontsize=14, fontweight='bold')

plt.title('Dataset GSE251778\nDistribución de muestras',
          fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Número de muestras', fontsize=12)
plt.ylim(0, 110)
plt.tight_layout()
plt.savefig('distribucion_muestras.png', dpi=150)
plt.show()
print("✓ Gráfico guardado")
```

```python
import pandas as pd

# Construir tabla de metadatos clínicos
metadatos = []

for nombre, muestra in gse.gsms.items():
    chars = muestra.metadata.get('characteristics_ch1', [])
    fila = {'sample_id': nombre}
    for c in chars:
        if ':' in c:
            clave, valor = c.split(':', 1)
            fila[clave.strip()] = valor.strip()
    metadatos.append(fila)

df_meta = pd.DataFrame(metadatos)
print("Tabla de metadatos:")
print(df_meta.head(10))
print(f"\nColumnas disponibles: {list(df_meta.columns)}")
df_meta.to_csv('metadatos_GSE251778.csv', index=False)
print("✓ Tabla guardada como CSV")
```

```python
# FASE 1 — Descarga del dataset de transcriptómica
import GEOparse

print("Conectando con NCBI GEO...")
gse = GEOparse.get_GEO(geo="GSE251778",
                        destdir="./datos/",
                        silent=False)

print("\n--- RESULTADO ---")
print("Título del estudio:", gse.metadata['title'][0])
print("Número de muestras:", len(gse.gsms))
```

```python
# Ver los primeros 5 pacientes del dataset
print("Primeras 5 muestras del estudio:\n")

for i, (nombre, muestra) in enumerate(gse.gsms.items()):
    if i >= 5:
        break
    titulo = muestra.metadata.get('title', ['Sin título'])[0]
    caracteristica = muestra.metadata.get(
        'characteristics_ch1', ['Sin info'])
    print(f"ID: {nombre}")
    print(f"Título: {titulo}")
    print(f"Características: {caracteristica[0]}")
    print("---")
```

```python
# Separar TDM de controles
controles = []
tdm = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]
    if 'Control' in titulo:
        controles.append(nombre)
    elif 'MDD' in titulo or 'Depression' in titulo or 'Case' in titulo:
        tdm.append(nombre)

print(f"Total muestras: {len(gse.gsms)}")
print(f"Controles sanos: {len(controles)}")
print(f"Pacientes TDM: {len(tdm)}")
print(f"\nPrimeros 3 controles: {controles[:3]}")
print(f"Primeros 3 TDM: {tdm[:3]}")
```

```python
import pandas as pd

# Crear tabla con información de cada muestra
info_muestras = []

for nombre, muestra in gse.gsms.items():
    titulo = muestra.metadata.get('title', [''])[0]

    # Determinar grupo
    if 'Control' in titulo:
        grupo = 'Control'
    else:
        grupo = 'TDM'

    # Determinar sexo
    if '-M-' in titulo:
        sexo = 'Masculino'
    elif '-F-' in titulo:
        sexo = 'Femenino'
    else:
        sexo = 'Desconocido'

    info_muestras.append({
        'ID_muestra': nombre,
        'Titulo': titulo,
        'Grupo': grupo,
        'Sexo': sexo
    })

# Convertir a tabla
df_muestras = pd.DataFrame(info_muestras)

# Mostrar resumen
print("=== RESUMEN DEL DATASET GSE251778 ===\n")
print(df_muestras.groupby(['Grupo', 'Sexo']).size())
print(f"\nTotal: {len(df_muestras)} muestras")

# Guardar como archivo CSV
df_muestras.to_csv('metadata_GSE251778.csv', index=False)
print("\n✓ Metadata guardada en metadata_GSE251778.csv")
```

```python
import pandas as pd

# Cargar la matriz de expresión génica
print("Cargando matriz de expresión génica...")
df_expresion = pd.read_csv(
    './datos/GSE251778/GSE251778_Mokhtari_rna_counts.csv.gz',
    compression='gzip',
    index_col=0
)

print(f"✓ Matriz cargada")
print(f"Genes totales: {df_expresion.shape[0]}")
print(f"Muestras totales: {df_expresion.shape[1]}")
print(f"\nPrimeros 5 genes:")
print(df_expresion.head())
```

```python
# Buscar IDO1 en los datos
# IDO1 tiene el código Ensembl: ENSG00000131203

print("=== BUSCANDO IDO1 ===\n")

# Buscar por código Ensembl
ido1_ensembl = 'ENSG00000131203'

if ido1_ensembl in df_expresion['gene_ID'].values:
    ido1_data = df_expresion[
        df_expresion['gene_ID'] == ido1_ensembl
    ]
    print("✓ IDO1 ENCONTRADO EN EL DATASET")
    print(f"Código Ensembl: {ido1_ensembl}")
    print(f"\nNiveles de expresión en primeras 10 muestras:")
    print(ido1_data.iloc[:, 1:11])

    # Estadísticas básicas
    valores = ido1_data.iloc[:, 1:].values.flatten()
    print(f"\nExpresión mínima: {valores.min()}")
    print(f"Expresión máxima: {valores.max()}")
    print(f"Expresión promedio: {valores.mean():.2f}")
else:
    print("IDO1 no encontrado con ese código")
    print("Buscando variantes...")
    resultados = df_expresion[
        df_expresion['gene_ID'].str.contains('ENSG00000131203',
                                              na=False)
    ]
    print(resultados)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extraer valores de IDO1
ido1_valores = df_expresion[
    df_expresion['gene_ID'] == 'ENSG00000131203'
].iloc[:, 1:]

# Identificar columnas de cada grupo
# Los IDs de muestra tienen el formato: 111_rna, 113_rna, etc.
muestras_cols = ido1_valores.columns.tolist()

# Separar por grupo usando el metadata que guardamos
tdm_ids = df_muestras[df_muestras['Grupo'] == 'TDM']['ID_muestra'].tolist()
ctrl_ids = df_muestras[df_muestras['Grupo'] == 'Control']['ID_muestra'].tolist()

# Obtener valores numéricos de IDO1
ido1_fila = ido1_valores.values.flatten()

# Calcular promedio por grupo (aproximación con posiciones)
n_total = len(ido1_fila)
mitad = n_total // 2

grupo1 = ido1_fila[:mitad]
grupo2 = ido1_fila[mitad:]

print("=== IDO1: COMPARACIÓN ENTRE GRUPOS ===\n")
print(f"Grupo 1 — Promedio IDO1: {grupo1.mean():.2f}")
print(f"Grupo 2 — Promedio IDO1: {grupo2.mean():.2f}")
print(f"Diferencia: {abs(grupo1.mean() - grupo2.mean()):.2f}")

# Gráfico
fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot([grupo1, grupo2],
           labels=['Grupo 1', 'Grupo 2'])
ax.set_title('Expresión de IDO1 entre grupos\nDataset GSE251778')
ax.set_ylabel('Nivel de expresión (conteos)')
ax.set_xlabel('Grupo')
plt.tight_layout()
plt.savefig('IDO1_expresion_grupos.png', dpi=150)
plt.show()
print("\n✓ Gráfico guardado")
```
