```python
import GEOparse

print("Descargando GSE251778...")
print("Esto puede tardar unos minutos...")

gse = GEOparse.get_GEO(geo="GSE251778",
                        destdir="./datos/",
                        silent=False)

print("\n✓ Dataset descargado")
print("Título:", gse.metadata['title'][0])
print("Número de muestras:", len(gse.gsms))
```
